## Line Charts

```python
years = range(2000, 2012)
apples = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931, 0.934, 0.936, 0.937, 0.9375, 0.9372, 0.939]
oranges = [0.962, 0.941, 0.930, 0.923, 0.918, 0.908, 0.907, 0.904, 0.901, 0.898, 0.9, 0.896, ]
```

```python
# Global API - linear construction, order matters
plt.figure(figsize= (8, 4))
plt.plot(years, apples, 's-g')      # fmt: marker - linetype - color
plt.plot(years, oranges, '.--r')    # point - double -, orange

plt.xlabel('Years')
plt.ylabel('Tons of product')
plt.title('Crob Yields')
plt.legend(['Apples', 'Oranges'])
```

```python
# OOP API - object construction
#https://matplotlib.org/stable/tutorials/introductory/quick_start.html#sphx-glr-tutorials-introductory-quick-start-py

fig, ax = plt.subplots(figsize=(8,4))   # instantiate two objects

ax.plot(years, apples, 's-g')      # fmt: marker - linetype - color
ax.plot(years, oranges, '.--r')    # point - double -, orange

ax.set_xlabel('Years')
ax.set_ylabel('Tons of product')
ax.set_title('Crob Yields')
ax.legend(['Apples', 'Oranges'])
```

### Multiple Histograms

```python
spcs_a = flowers_df.species.unique()
spcs_a
```

```python
spcs_lst = []
for spc in spcs_a:
    spcs_lst.append(flowers_df[flowers_df.species == spc])
#spcs_lst  # each elem is a diff specie DF
```

```python
# setosa_df = flowers_df[flowers_df.species == 'setosa']
# versicolor_df = flowers_df[flowers_df.species == 'versicolor']
# virginica_df = flowers_df[flowers_df.species == 'virginica']
```

```python
fig, ax4 = plt.subplots()
bins_to, alpha_to = np.arange(2, 5, 0.25), 0.4
for i in range(len(spcs_lst)):
    ax4.hist(spcs_lst[i].sepal_width, alpha= alpha_to, bins= bins_to)
ax4.legend(spcs_a)

```

```python
# Multiple Histogram Stacked
fig, ax4 = plt.subplots()
ax4.hist([spcs_lst[0].sepal_width, spcs_lst[1].sepal_width,
          spcs_lst[2].sepal_width], bins= bins_to, stacked=True)
ax4.legend(spcs_a)
```

## Heatmap graphs
Visualize a 2D data like a matrix or a table

```python
flights_df = sns.load_dataset('flights')
flights_df
# This is not a 2D data set 
```

```python
# some quick graph we can do..
plt.plot(flights_df.passengers)
```

```python
# Convert de DF in a 2D dataset
flights_df_2d = flights_df.pivot('month', 'year', 'passengers')
flights_df_2d
```

```python
# now we can use hetmap
fig, ax9 = plt.subplots(figsize=(10,4))
ax9.set_title('No. of Passenger')
ax9 = sns.heatmap(flights_df_2d)
```

```python
fig, ax9 = plt.subplots(figsize=(12, 5))
ax9.set_title('No. of Passenger (1000s)')
ax9 = sns.heatmap(flights_df_2d, fmt='d', annot=True, cmap='Blues')
```
