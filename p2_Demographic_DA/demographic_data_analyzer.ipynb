```python
# 222 - What is the average age of men?
print(df['age'].mean())                         # of course same as in df.describe(): 38.581647
fam = df[df['sex'] == 'Female']['age'].mean()   # female age mean (average age)
mam = df[df['sex'] == 'Male']['age'].mean()     # male age mean (average age)
am = (fam + mam) / 2                            # same as age.mean?
print(fam, mam, am)
print(df.loc[df['sex'] == 'Male']['age'].mean())
df[df['sex'] == 'Male']['age'].mean()
# print(df['age'].mean())
# #df.loc[df['sex'] == 'Male']['age']
# #df.loc[df['sex'] == 'Female']
# print(df[(df['race'] == 'Black')])
```

```python
# 333 - What is the percentage of people who have a Bachelor's degree?
#df[df['education'] == 'Bachelors']
print(df.value_counts(df['education'] == 'Bachelors'))
#print(df['education'].value_counts())
pwb = df[df['education'] == 'Bachelors'].shape[0]       # people w/ Bachelors
tp = df.shape[0]                                        # total people
pwb_percent = pwb / tp * 100
print(pwb, tp, pwb_percent)  
#df['education'].value_counts(normalize=True) * 100
print(df['education'].value_counts(normalize=True)['Bachelors'] * 100)

# Other --use count--!
(df['sex'].value_counts()/df['sex'].count())*100
```

```python
# 444 - What percentage of people with advanced education (Bachelors, Masters, or Doctorate)
# make more than 50K?

# print(df.value_counts(df['education']))   # 1st education categories
# print(df[(df['education'] == 'Bachelors') | (df['education'] == 'Masters')  | (df['education'] == 'Doctorate')])    # second separate Bachelors, Masters, Doctorate
pae = df[(df['education'] == 'Bachelors') | (df['education'] == 'Masters')  | (df['education'] == 'Doctorate')]     # this a a new (smaller) DF: people advance education
#display(pae)
#print(pae.info())
#pae[pae['salary'] == '>50K']  # filter salary >50K
print(pae['salary'].value_counts(normalize=True) * 100)           # i got a Serie w/ percentage of each category (catetg is the series index)
pae_salary_serie = pae['salary'].value_counts(normalize=True) * 100
print(type(pae_salary_serie), '  -  ', pae_salary_serie.index)

pae['salary'].value_counts(normalize=True)['>50K'] * 100    # get the percentage >50 for advance educ people
```

```python
# 555 - What percentage of people without advanced education make more than 50K?
p_nae = df[~((df['education'] == 'Bachelors') | (df['education'] == 'Masters')  | (df['education'] == 'Doctorate'))]
#p_nae.value_counts(pnae['education'])        # as you can see NOT advance education people here.
p_nae['salary'].value_counts(normalize=True)['>50K'] * 100    # I ger the percentage >50 for adv. educ people
```

```python
# 666 - What is the minimum number of hours a person works per week?
print(type(df['hours-per-week'].min()))
df['hours-per-week'].min()      # of course same value as in df.describe()
```

```python
# 777 - What percentage of the people who work the minimum number of hours per week have
# a salary of more than 50K?

#df[df['hours-per-week'] == df['hours-per-week'].min()]    # all people that work the min number h/week
pw_min = df.loc[df['hours-per-week'] == df['hours-per-week'].min()]     # new DF w/people who work the minimum number of hours per week
print(pw_min.shape[0])
#pw_min = df[df['hours-per-week'] == 1]
#display(pw_min)
#print(type(pw_min))
print(pw_min.shape)
print(pw_min.value_counts(pw_min['salary']) )   # plain value counts of elements of each category in salary
print(pw_min['salary'].value_counts())          # same as above
print(pw_min['salary'].value_counts(normalize=True) * 100)          # same as above but normalize (relation) + * 100 to percentage
# as always: the above es a Series and first column is the index.
#pw_min['salary'].value_counts(normalize=True)['>50K'] * 100
df.loc[df['hours-per-week'] == df['hours-per-week'].min()] ['salary'].value_counts(normalize=True)['>50K'] * 100
```
