### Step 2: Load the data
- into a pandas DataFrame (df)
- we'll use fifa.csv dataset

```python
import pandas as pd
import zipfile as zfm       # my datasets are zipped csv files

with zfm.ZipFile('DataToVisualize.zip') as zf:
    df = pd.read_csv(zf.open('fifa.csv'), index_col='Date', parse_dates=True)

df.head(2)
```

### Step 3: Plot the data
- Now that the data is loaded into the notebook, you're ready to visualize it!
- I use OOP style

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, ax1 = plt.subplots(figsize=(16,6))
sns.lineplot(data=df, ax=ax1)
ax1.set_title('Historical FIFA rankings for: ARG, BRA, ESP, FRA, GER, ITA')
```
