## 5. Distributions
- Create histograms and density plots

```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
'Setup Complete'
```

### Selecting the dataset

```python
zipfile = 'DataToVisualize.zip'
dataset = 'iris.csv'
```

```python
# Inspecting the zip file
import zipfile as zfm

with zfm.ZipFile(zipfile) as zf:
    if dataset in zf.namelist():
        print(f'Datataset "{dataset}" founded in "{zipfile}"')
```

```python
# Inspecting insurance.csv
with zfm.ZipFile(zipfile) as zf:
    #print(zf.namelist())
    numln = 1
    for ln in zf.open(dataset):
        print(ln)
        if numln == 5: break
        numln += 1
```

### Load & Examine the Data

```python
with zfm.ZipFile(zipfile) as zf:
    df = pd.read_csv(zf.open(dataset))
print(df.shape)
#df.describe()
df.iloc[[0, 1, 9, -9, -2, -1]]
```

```python
### set Id as df.index
df.set_index('Id', inplace=True)
print(df.shape)
df.iloc[[0, 1, -2, -1]]
```

### Histogram
- ex. how petals lenght varies in iris flowers

```python
fig, axhist = plt.subplots(figsize=(5,3))
sns.histplot(df['Petal Length (cm)'], ax=axhist)
```

### Density plots
- a KDE plot (Kernel Density Estimate)  |  .kdeplot()

```python
fig, axhist = plt.subplots(figsize=(5,3))
sns.kdeplot(df['Petal Length (cm)'], fill=True, ax=axhist)
```

### 2D KDE plots
- .joinplot()

```python
# jointplot is a figure-level function. -> high + aspect
# sns.jointplot(x=df['Petal Length (cm)'], y=df['Sepal Width (cm)'],
#               fill=True, kind='kde')
sns.jointplot(x=df['Petal Length (cm)'], y=df['Sepal Width (cm)'],
              kind='kde')
```

### Color-coded plots
#### histogram separated by species

```python
sns.histplot(data=df, x='Petal Length (cm)', hue='Species')
```

#### KDE colored plots (hue by species)

```python
fig, axkde1 = plt.subplots(figsize=(5,3))
sns.kdeplot(data=df, x='Petal Length (cm)', hue='Species',
             fill=True, ax=axkde1)
axkde1.set_title("Distribution of Petal Lengths, by Species")
```

```python
## lets take a look how many of each Specie
df.Species.value_counts()
```

```python
dfversi = df[df.Species == 'Iris-versicolor']
#dfversi.reset_index(inplace=True)
dfversi.iloc[[0, 4, 9, -9, -4, -1]]
```

```python
# Histogram (histplot)
plt.figure(figsize=(4,3))
sns.histplot(dfversi['Petal Length (cm)'])
```

```python
# Density plot (KDE plots)
plt.figure(figsize=(4,3))
sns.kdeplot(dfversi['Petal Length (cm)'], fill=True)
```

```python
# quick statistics view
df.describe()
```

- mean is the average value
- median is a middle value when data es orderer
- mode is the value tha occurs more often
> For most practical purposes, lower quartile, median, and upper quartile are the 25th, 50th and 75th percentiles, respectively.

```python
# .mean(), .median(), .mode()
mean = df['Petal Length (cm)'].mean()
median = df['Petal Length (cm)'].median()
mode = df['Petal Length (cm)'].mode().iloc[0]
mode
```

```python
# KDE plot with statistics
ymi, ymx = 0, 1

fig, axkdes = plt.subplots(figsize=(6,4))
sns.kdeplot(dfversi['Petal Length (cm)'], ax=axkdes)
axkdes.vlines(x=mean, colors='r', label='mean', ymin=ymi, ymax=ymx)
axkdes.vlines(x=median, colors='g', label='median', ymin=ymi, ymax=ymx)
axkdes.vlines(x=mode, colors='y', label='mode', ymin=ymi, ymax=ymx)
axkdes.set_title('Petal length distribution for iris versicolor')
axkdes.legend()
```
