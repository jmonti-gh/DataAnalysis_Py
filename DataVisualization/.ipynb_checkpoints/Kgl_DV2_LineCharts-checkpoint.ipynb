## 2. Line Charts
- Visualize trends over time

```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
'Setup Complete'
```

### Selecting the dataset

```python
import zipfile as zfm

with zfm.ZipFile('DataToVisualize.zip') as zf:
    numln = 1
    for ln in zf.open('spotify.csv'):
        print(ln)
        if numln == 15: break
        numln += 1
```

### Load & Esamine the Data

```python
with zfm.ZipFile('DataToVisualize.zip') as zf:
    df = pd.read_csv(zf.open('spotify.csv'), index_col='Date', parse_dates=True)
df
```

### Plot the data

```python
fig, axe = plt.subplots(figsize=(16,6))
sns.lineplot(data=df, ax=axe)
axe.set_title('Daily Global Streams of Popular Songs in 2017-2018')
```

### Plot a subset of data

```python
df.columns, type(df.columns)
```

```python
cols = list(df.columns)
cols, type(cols)
```

```python
### plot the lines corresponding to the first two columns
fig, ax1 = plt.subplots(figsize=(16,6))
sns.lineplot(data=df['Shape of You'], label='Shape of you', ax=ax1)
sns.lineplot(data=df['Despacito'], label='Despacito', ax=ax1)
#ax1.set_xlabel('Date')
ax1.set_xlabel(df.index.name)
ax1.set_title('Daily Global Streams of "Shape of You" and "Despacito" in 2017-2018')
```

```python
df[['Shape of You', 'Despacito']]
```

```python
### Using matplotlib.pyplot
fig, axm = plt.subplots(figsize=(16,6))
axm.plot(df.index, df[['Shape of You', 'Despacito']])
```

```python
fig, axm = plt.subplots(figsize=(16,6))
axm.plot(df.index, df['Shape of You'], label='Shape of You')
axm.plot(df.index, df['Despacito'], label='Despacito')
axm.legend()
axm.set_title('Daily Global Streams of Popular Songs in 2017-2018')
```
