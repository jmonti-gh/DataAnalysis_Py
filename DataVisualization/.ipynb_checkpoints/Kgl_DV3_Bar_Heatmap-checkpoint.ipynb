## 2. Bar Charts and Heatmaps
- Use color or length to compare categories in a dataset

```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
'Setup Complete'
```

### Selecting the dataset

```python
import zipfile as zfm

with zfm.ZipFile('DataToVisualize.zip') as zf:
    #print(zf.namelist())
    numln = 1
    for ln in zf.open('flight_delays.csv'):
        print(ln)
        if numln == 5: break
        numln += 1
```

### Load & Examine the Data

```python
with zfm.ZipFile('DataToVisualize.zip') as zf:
    df = pd.read_csv(zf.open('flight_delays.csv'), index_col='Month')
print(df.shape)
#df.describe()
df
```

### Bar chart

```python
### bar chart showing the average arrival delay for Spirit Airlines
### (airline code: NK) flights, by month.

fig, axbar = plt.subplots(figsize=(12,6))
sns.barplot(x=df.index, y=df.NK, ax=axbar)
axbar.set_title('Average Arrival Delay for Spirit Airlines Flights, by Month')
axbar.set_ylabel('Arrival delay (in minutes)')
```

### Heatmap

```python
fig, axheat = plt.subplots(figsize=(14,7))
sns.heatmap(data=df, annot=True, ax=axheat)
axheat.set_title('Average Arrival Delay for Each Airline, by Month')
axheat.set_xlabel('Airline')
```

### to see nulls (NaN) values

```python
### a general view
df.info()
```

```python
### columns that has null values
df.isnull().sum()
```

```python
### fill nulls using pandas
df.US.fillna(value=df.US.mean())
```

```python
### mk a new df to see diff fillna values and methods
fna_df = pd.DataFrame()

# put cols to fillna_df
fna_df['OriginalCol'] = df.US

fna_df
```

```python
# fillna with average + median + bfill + ffill
fna_df['w/mean'] = df.US.fillna(df.US.mean())
fna_df['w/median'] = df.US.fillna(df.US.median())
fna_df['w/bfill'] = df.US.fillna(method='bfill')
fna_df['w/ffill'] = df.US.fillna(method='ffill')

fna_df
```
