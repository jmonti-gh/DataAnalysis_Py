### 1.0.- Import libraries and read zipfile & dataset content

```python
import zipfile as zfm
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

zipfile = 'BankCustChurn.zip'
dataset = 'Bank Customer Churn Prediction.csv'

with zfm.ZipFile(zipfile) as zf:
    df = pd.read_csv(zf.open(dataset), index_col='customer_id')
print(df.shape)
df.iloc[[0, 5, 9, -9, -5, -1]]
```

#### 2.1.- Correlation Heatmap (correlogram)
- https://vitalflux.com/correlation-heatmap-with-seaborn-pandas/
- Correlation is a statistical measure that expresses the strength of the relationship between two variables.
- Correlation heatmaps are a type of plot that visualize the strength of relationships between numerical variables. 
- https://towardsdatascience.com/all-about-heatmaps-bb7d97f099d7
- https://stackabuse.com/ultimate-guide-to-heatmaps-in-seaborn-with-python/

```python
# Build correlation pandas DataFrame
corr = df.corr(numeric_only=True)
# Generate a mask for upper traingle
# (numpy.triu(m, k=0) u -> Upper triangle of an array; tril l -> Lower)
mask = np.triu(np.ones_like(corr, dtype=bool))
# Configure a custom diverging colormap if you want different than default
cmap = sns.diverging_palette(230, 20, as_cmap=True)
# Set up the matplotlib plot configuration
fig, axcorr = plt.subplots(figsize=(10,6))
sns.heatmap(corr, annot=True, linecolor='g', linewidth=0.5, ax=axcorr)
#sns.heatmap(corr, annot=True, mask=mask, cmap=cmap, ax=axcorr)
```
