## Choosing Plot Types and Custom Styles
> Customize your charts and make them look snazzy.    
- Trends (pattern of change - over a period of time by ex.)
    - Line charts: __sns.lineplot(...)__, are best to show trends over a period of time, and multiple lines can be used to show trends in more than one group. 
 - Relationship (between variables in the data)
     1. Bar charts: sns.barplot(...), are useful for comparing quantities corresponding to different groups.
     2. Heatmaps: sns.heatmap(...), can be used to find color-coded patterns in tables of numbers.
     3. Scatter plots: sns.scatterplot(...), show the relationship between two continuous variables; if color-coded (hue=), we can also show the relationship with a third categorical variable.
     4. Regression line: sns.regplot(...), Including a regression line in the scatter plot makes it easier to see any linear relationship between two variables.
     5. Mult. Regr. lines: sns.lmplot(...), This command is useful for drawing multiple regression lines, if the scatter plot contains multiple, color-coded groups. (JM try w/ no color-coded)
     6. Categ. Scatt. plots: __sns.swarmplot(...)__, show the relationship between a continuous variable and a categorical variable.
- Distribution (possible values in a variable, along with how likely they are)
    1. Histograms: __sns.histplot(...)__, show the distribution of a single numerical variable. (Bars)
    2. KDE plots: __sns.kdeplot(...)__, (or 2D KDE plots) show an estimated, smooth distribution of a single numerical variable (or two numerical variables). -Kernel Density Estimate-
    3. 2D KDE plots: __sns.jointplot(...)__, This command is useful for simultaneously displaying a 2D KDE plot with the corresponding KDE plots for each individual variable.

## Changing styles with seaborn

```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

### Selecting the dataset

```python
zipfile = 'DataToVisualize.zip'
dataset = 'spotify.csv'
```

```python
# Inspecting the zip file
import zipfile as zfm

with zfm.ZipFile(zipfile) as zf:
    if dataset in zf.namelist():
        print(f'Datataset "{dataset}" founded in "{zipfile}"')
```

```python
# Inspecting insurance.csv
with zfm.ZipFile(zipfile) as zf:
    #print(zf.namelist())
    numln = 1
    for ln in zf.open(dataset):
        print(ln)
        if numln == 5: break
        numln += 1
```

### Load & Examine the Data

```python
with zfm.ZipFile(zipfile) as zf:
    df = pd.read_csv(zf.open(dataset))
    # spotify_data = pd.read_csv(spotify_filepath, index_col="Date", parse_dates=True)
print(df.shape)
#df.describe()
df.iloc[[0, 1, 9, -9, -2, -1]]
```

```python
### set Id as df.index
df.set_index('Date', inplace=True)
print(df.shape)
df.iloc[[0, 1, -2, -1]]
```

```python
### Change index dtype, 'O'bject (str) -> 'DateTime'
print('Index original dtype:', df.index.dtype)    # dtype('O') - df.index.dtype
df.index = pd.to_datetime(df.index)
print('Index actual dtype:', df.index.dtype)     # dtype('<M8[ns]') - df.index.dtype
```

### Line Chart - .lineplot() - axes_level_plot -

```python
fig, axline = plt.subplots(figsize=(9,5))
sns.lineplot(df, ax=axline)
```

```python
# Change the style of the figure to the "dark" theme
sns.set_style("dark")

# Line chart 
plt.figure(figsize=(12,6))
sns.lineplot(data=df)
```

> Seaborn has five different themes: (1)"darkgrid", (2)"whitegrid", (3)"dark", (4)"white", and (5)"ticks", and you need only use a command similar to the one in the code cell above (with the chosen theme filled in) to change it.
