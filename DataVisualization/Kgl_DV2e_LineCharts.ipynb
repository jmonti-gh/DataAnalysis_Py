### Step 1: Load and review the data
- into a pandas DataFrame (df)
- we'll use museum_visitors.csv dataset

```python
with zfm.ZipFile('DataToVisualize.zip') as zf:
    df = pd.read_csv(zf.open('museum_visitors.csv'), index_col='Date',
                     parse_dates=True)

df.tail(5)
```

```python
# Fill in the line below: How many visitors did the Chinese American Museum 
# receive in July 2018?
ca_museum_jul18 = df.loc['2018-07-01', 'Chinese American Museum']
print(f'{ca_museum_jul18:,}')
print(df.iloc[25, 0], '  -  ', df.index[25])

# Fill in the line below: In October 2018, how many more visitors did Avila 
# Adobe receive than the Firehouse Museum?
date = '2018-10-01'
avila_oct18 = df.loc[date, 'Avila Adobe'] -df.loc[date, 'Firehouse Museum']
print(f'{avila_oct18:,}')
```

### Step 2: Convince the museum board
To show the museum board how the event compared to regular traffic at each museum, create a line chart that shows how the number of visitors to each museum evolved over time. Your figure should have four lines (one for each museum).

```python
fig, ax1 = plt.subplots(figsize=(16,6))
sns.lineplot(data=df, ax=ax1)
ax1.set_title('Museum monthly visitors')
```

### Step 3: Assess seasonality
low seasons (when the employees are perfectly staffed and happy) and also high seasons (when the employees are understaffed and stressed)
#### Part A
Create a line chart that shows how the number of visitors to Avila Adobe has evolved over time.

```python
fig, ax2 = plt.subplots(figsize=(16,6))
sns.lineplot(data=df['Avila Adobe'], ax=ax2)
ax2.set_title("Monthly Visitors to Avila Adobe")
ax2.set_xlabel(df.index.name)
```
