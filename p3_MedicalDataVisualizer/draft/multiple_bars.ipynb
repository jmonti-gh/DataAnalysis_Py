# Multiple Bars - matplotlib  & seaborn
- Grouped Bart chart
- Multiple bar plots, stacked bar plots, horizontal bar charts
- Multiple bar charts are generally used for comparing different entities.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mtl
import seaborn as sns

print('np:', np.__version__)
print('pd:', pd.__version__)
print('mtl:', mtl.__version__)
print('sns:', sns.__version__)
```

```python
# Data definition

X = ['Group A','Group B','Group C','Group D']
Ygirls = [10,20,20,40]
Zboys = [20,30,25,30]

Y2g = list(map(lambda x: x * 2, Ygirls))
Z2b = list(map(lambda x: x * 1.5, Zboys))
  
X_axis = np.arange(len(X))
```

```python
# Simple multiple bar chart
fig, axe = plt.subplots(figsize=(10, 6))

axe.set_title('Title')
axe.set_xlabel('x_label')
axe.set_ylabel('y_label')

axe.bar(X_axis - 0.2, Ygirls, 0.4, label = 'Girls')
axe.bar(X_axis + 0.2, Zboys, 0.4, label = 'Boys')
axe.legend()
axe.set_xticks(X_axis, X )
axe.spines['right'].set_visible(False)
axe.spines['top'].set_visible(False)
```

## MBC Side by Side

```python
fig, axs = plt.subplots(1, 2, figsize=(10, 6))     # (rows, cols, size of full fig)

x_labels, y_label = 'variable', 'total'

axs[0].set_title('Tit  = 0')
axs[0].set_xlabel(x_labels)
axs[0].set_ylabel(y_label)
axs[0].bar(X_axis - 0.2, Ygirls, 0.4, label = 'Girls')
axs[0].bar(X_axis + 0.2, Zboys, 0.4, label = 'Boys')
#axs[0].legend()
axs[0].set_xticks(X_axis, X )
axs[0].spines['right'].set_visible(False)
axs[0].spines['top'].set_visible(False)

axs[1].set_title('Tit = 1')
axs[1].set_xlabel(x_labels)
axs[1].bar(X_axis - 0.2, Y2g, 0.4, label = 'Girls')
axs[1].bar(X_axis + 0.2, Z2b, 0.4, label = 'Boys')
axs[1].legend(loc='center right')
axs[1].set_xticks(X_axis, X )
axs[1].spines['right'].set_visible(False)
axs[1].spines['top'].set_visible(False)
```
